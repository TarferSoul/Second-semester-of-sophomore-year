```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a function to draw the workflow diagram
def draw_workflow_diagram(title, steps, color):
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 10)
    ax.axis('off')

    # Draw arrows and boxes
    for i, (x, y, text) in enumerate(steps):
        rect = mpatches.FancyBboxPatch((x-1, y-0.5), 6, 1.5, boxstyle="round,pad=0.3", edgecolor='black', facecolor=color, mutation_scale=0.2)
        ax.add_patch(rect)
        ax.text(x+2, y+0.25, text, ha='center', va='center', fontsize=10, wrap=True)
        if i > 0:
            ax.arrow(x+2, y+1.25, 0, 0.75, head_width=0.2, head_length=0.2, fc='black', ec='black')
    
    ax.text(1, 9, title, fontsize=14, fontweight='bold', ha='left')
    plt.show()

# Special Case Steps
steps_special = [
    (2, 7.5, "Select Pre-trained Models:\nResNet18, ResNet34, ResNet50"),
    (2, 6, "Apply Bottom Layer Freezing"),
    (0, 4.5, "ResNet18:\nFreeze layers 29-40"),
    (2, 4.5, "ResNet34:\nFreeze layers 49-72"),
    (4, 4.5, "ResNet50:\nFreeze layers 87-106"),
    (2, 2, "Fine-tune Models for CIFAR100's\n100 Downstream Classification Services")
]

# General Case Steps
steps_general = [
    (2, 7.5, "Select Pre-trained Models:\nResNet18, ResNet34, ResNet50"),
    (2, 6, "Full Parameter Fine-tuning on CIFAR100\nwith Selected Superclasses"),
    (2, 4.5, "Apply Bottom Layer Freezing\nand Fine-tune on Sub-datasets"),
    (2, 3, "Create Shared Parameter Blocks"),
    (2, 1.5, "Second Round Fine-tuning for\nEach Class within Every Superclass")
]

# Draw the diagrams
draw_workflow_diagram('Special Case Workflow Diagram', steps_special, 'orange')
draw_workflow_diagram('General Case Workflow Diagram', steps_general, 'lightblue')
```
